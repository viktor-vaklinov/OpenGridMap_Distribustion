```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
```

```python
import random
G=nx.complete_graph(5)
for e in G.edges(data=True):
    e[2]['weight'] = random.uniform(1,10)
    e[2]['capacity'] = 2
G=G.to_directed()
for n in G.nodes(data=True):
    n[1]['demand'] = 1
    G.add_edge('S',n[0],weight= random.uniform(1,10))

G.add_node('S', demand = -5)
# print G.edges(data=True)
# print G.nodes(data=True)

flowCost, flowDict = nx.capacity_scaling(G)#network_simplex(G)

nx.draw_networkx(G,pos=nx.spring_layout(G),) 
plt.show()
    
```

```python
ss = nx.minimum_branching(G)
nx.draw_networkx(ss,pos=nx.spring_layout(ss),) 
plt.show()
```

```python
eList = sorted([(u, v) for u in flowDict for v in flowDict[u] if flowDict[u][v] > 0])
#print eList, G.nodes()
G_ = nx.DiGraph()
G_.add_node(1)
print type(G_)
G_.add_nodes_from([0, 1, 2, 3, 4, 'S'])
G_.add_edges_from(eList)

nx.draw_networkx(G_,pos=nx.spring_layout(G_),) 
plt.show()
```
